## Data Cleaning

### Removing Junk Values

```python
pattern = r'[a-zA-Z0-9!@#$%^&*()_+{}\[\]:;<>,.?~|"\']+$'
data = data[data['course_title'].str.contains(pattern,regex = True)]
```

```python
data.shape
```

### 3. Courses wich are free of cost

### 4.Courses which are Paid

```python
data['is_paid'] = data.is_paid.astype(str)
```

```python
data[data['is_paid'] == 'True']
```
